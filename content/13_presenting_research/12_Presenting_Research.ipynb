# Presenting your Research

Sharing and discussing research with a group can be one of the best experiences in science. In the best presentations, you are carefully brought along on the path that a team of fellow scholars has walked for many months or years, and get to see things that are strange or fascinating along the way. You leave feeling that what you knew already you now know better; or that something you thought was so simple turns out to hold depths you'd never imagined; or that something you thought was obvious turns out not to be right at all. You feel that each choice that the presenter made - from the design of slides to the way they move - has been considered with you in mind. The best presentations are gracious to past research, and show how lines of work connect or challenge one another. 

The worst presentations bowl you


This will be a 12 minute presentation of your groups work on the project in class, in the format of a scientific presentation. Here are my thoughts on making presentations awesome. I don’t always do or live up to all of them, but I am pretty careful to do all of these for presentations where I want to make a good impression. I don’t have time to do them all I usually at least know what I might have problems with and try to compensate as best I can (e.g. diagram on whiteboard or with arms if I haven’t had time to make a diagram slide).  

#### Presentation length vs. presentation time
It’s important to have your presentation at about the right length. Speakers vary a lot in terms of how much material they can present clearly in a given amount of time, and in their preferences.
Here's a general formula you can use:

As a maximum, present no more than 1 slide per 30 seconds (so 24 slides total for a 12 minute talk, 20 slides for a 10 minute talk, up to 40 for a 20 minute talk, etc). 

This is a fairly rapid pace which takes some practice to get smooth and not sound rushed while also conveying all the needed info.  Note that if you use slides for transitions (e.g. 5 slides each introducing one series on your graph to make it easier to explain), you may have a slightly higher number...but still no more than 2* length of talk in minutes 'real' slides. This timing also implies you are saying just one fairly simple main thing per slide with no more than about three supporting statements. At this pace you do still have time to walk folks through figures carefully, but can't get lost in tangents.

**Note:** If you don’t have time to rehearse or break slides up into manageable parts 1) make time to do that! 2) you probably will need closer to 1 minute per slide, and you have to be careful to be disciplined about what exactly you want to say.

On the other end, I wouldn’t present a lot fewer than 1 slide per minute on average unless you have some very profound slides that each require a lot of discussion and deliberation. (Even then, maybe break it up a little to channel that discussion and deliberation)


Breakdown of slides
Let us consider as an example a 15 minute talk. Allowing for at least 3 minutes of questions, we have only 12 minutes to work with.  Here’s how I might think about breaking down the slides. Your mileage may vary – but be very sure to include the relevant background, hypotheses, and methods.

- 3-4 slides.  **Background and Significance (see below)**.  Why should we care about your problem?  What's interesting about it?  What are the general terms/concepts we need to know to understand your problem? What is already known and how do you build on that.
- 1-2 slides. **Hypothesis**. Represent alternative scenarios visually if at all possible.
- 1-2 slide Approach. Clear description and diagram of the analysis used to address it. Workflow diagrams that are clearly explained work well here. 
- 1-2 slides. **Predicted results by hypothesis**. What results you expect to see under each major hypothesis (show in the same format you will present your results, but you can draw in cartoon form and without units to make clear it’s a hypothetical)
- Results sections
    I tend to break each general type of result up into its own result section. For each result section, you can want to clearly explain what the motivation for collecting. Generally this means explaining why they test the hypothesis or how they follow from the previous set of results.
    - 1 slide. Motivation for collecting these results. (What prediction of the main hypothesis do they test, or what aspect of the central question do they answer?)
    - 2-3 slides actual results. Be sure to really explain *every* figure shown.
    - 1 slide. Implications of this set of results. Relate results back to your core hypotheses. What did they tell you about your hypotheses? Can you now rule some out? What questions remain open that might need another se t of results to answer?
    You can repeat this "sandwich" motivation-results-implications structure as many times as you need to in order to explain your results. Be sure to explain at the end of each section *why* the next set of results is the next logical step. 
    
- 1 slide.  Implications of Results 1.  Relate results back to core hypothesis. relate results back to your core hypotheses. What did they tell you about your hypotheses? Can you now rule some out?
- Additional results sections

If you have more results, you usually iterate the last 4 steps in about 6 slides: justify an analysis, describe the workflow, describe what you expected, describe what you got, describe what it means in a broader sense.  Depending on timing and how obvious your subsequent analyses are you may be able to cut some of the justification, but think carefully about whether the result will be clear.

If results are similar or the audience is familiar with the techniques and the implications are super obvious, you can sometimes get away with just 1-2 slides.  Example [comments in brackets]: “1. Using the same procedure, we also searched all the bat genomes [if you just explained an analysis looking at genetics of echolocation in whales] 2. We see the same patterns again! [if you did, or ‘we see a totally different pattern!’] 3. This suggests consistent links between gene content and echolocation in convergently evolved echolocation systems [or this suggests fundamentally different genetic mechanisms came into play each time echolocation evolved].” 
1 slide. Conclusion. conclusions (with pictures or icons if possible, recalling past results that support that conclusion).
1 slide. Acknowledgements. If you got help or data from other groups, a nice diagram from a friend, or help with a tricky command from the authors of a paper you contacted by e-mail, this would be the place to say it. For class presentations, you don’t need to spend time acknowledging the instructor, just put up your slide. Also, for very short talks, you don’t have to walk through all these.

If you use all the maximums, that’s about 16 slides + 2-6 * extra results sections.  So you can have 3 very short results sections total (about 16 + 2 + 2 = 20) or 2 slightly longer results sections (16 + 6 = 22) in a 10-12 minute talk, and stay under 1 slide per 30 seconds.  

2-3 Slides. Sneaky bonus answer slides. If in practice you often get the same questions, you can include a 1-slide piece of key data/ analysis to answer them after acknowledgements.  If you get the question then go to that slide to support your answer.  This is an especially useful way to smuggle in one extra key result in short presentations where there are very predictable questions the audience will ask about your research.

#### Practice! 

Paradoxically, short talks require the most practice.  Hour-long talks are relatively easy and you can get away with a lot more. OTOH, you can practice a 10 minute talk 6 times in an hour, and a 20 minute talk 3 times in an hour so with a couple hours you can actually run through your presentation a LOT.  Use these guidelines to help refine your presentation.

#### Cite literature!

It is usually not that much more work to pull example diagrams or literature results out of your citations rather than from random websites online, and can add a lot of credibility.  Folks typically cite the reference in the lower right corner in about size 10 font. Reading the associated papers is one of the best ways to have a good answer to audience questions.

#### Don’t be afraid to establish the basics.
If you’re talking about the antigenicity of proteins based on their number of transmembrane domains, say clearly what an antigen is, what a transmembrane domain is, that the part that goes through the membrane is hydrophobic etc. This is a great chance to cite and use diagrams from the literature to show you have a strong basic foundation in your area.  In very short presentations (10 minutes) you have to be a little choosy, but in 20 minute or longer presentations you can spend longer than you probably think laying out the basics.  Audiences tend to really like this, because even if you mostly remember, you banish any doubts that they have about what you mean, and establish that you are on a common terminological footing.

#### Get feedback from peers. 
Try giving the presentation to someone who might be in bio in general but not familiar with your project. Ask what was clear and what was confusing.  Listen to what questions they have and see if you can change your descriptions to make those confusing points clear the first time around.

#### Be sure to justify and explain the background for your problem. 
The temptation is to just dive in to your analysis, but usually you need a significant amount of background to get your audience to understand and buy in to your project. In a short presentation there is less time, which means you'll have to think very strategically about what your readers know and believe and what additional information/insights they'll need to care about your projects.For a very short talk, I'd likely cut down to just the most important 3-4 of these.  Your pieces of information will depend on your project.  
Some examples of what I might put on 4 slides to intro certain projects (just some ideas):

Coral microbiomes across coral diversity: 1. Most of life's diversity is microbial (with citations to e.g. Pace 1997 and tree showing metazoans branching from within microbial eukarotes). One implication is that animals evolved in the context of microbial diversity... and have a staggering range of symbioses with animals. 2. corals have microbes on 'em (cite literature, show diagram/result/picture from literature to highlight point). 3. These are known to play a key role in metabolism (Symbiodinium example) and defense (coral mucus bacteria).and play a role in response to environmental stressors like global warming (image, laboratory study results). 4. But corals are diverse (graphs comparing divergence times for mammals, flowering plants, corals), and we know nothing about bacterial symbionts in many species of corals. 5) ...move into our key hypotheses, questions, etc about diversity of coral-bacterial symbiosis.

Gene transfer 1. what is gene transfer, and basic mechanisms (big diagram) 2. happens all the time in nature (cite literature, show cool figure from paper quantifying) 3) super important for e.g. spread of antibiotic resistance (data from literature, citation in lower right) but also other traits 4) transition to your question and approach

Echolocation 1. What is echolocation/diagram of how it works 2. What animals do it/how is it distributed on the tree? 3. genes involved 4. transition to your analysis of which genes matter.
etc etc.

#### Use images and diagrams to convey information where possible

1st drafts inevitably have walls of bullet pointed text, which is not ideal. Retention studies show that having blocks of text and someone describing the results in similar but non-identical words results in lower retention than either text or verbal description alone. You want people to pay attention to you, so keep all text on the slide succinct.

That said it's OK to include e.g. method parameters/details of genomes analyzed on the slide so folks who are interested can check them out. So if you did a BLAST search, you might say you did a BLAST search and the e-value, but leave all the other parameters at the bottom of a slide (that still has a nice visual diagram/image for it's main focus) so folks could check 'em out if desired.

 Again, try to have a main diagram, image or result that clearly conveys your main point be the visual focus of each slide If you were testing a hypothesis, clearly describe the hypothesis before talking about your analysis and its results.  If you were developing a tool to test a hypothesis, describe what it was designed to test before describing the workflow.  In either case, be sure to relate back to your intro material on why this matters.

If you can, make a diagram of some kind that describes your hypothesis, rather than just writing it out in text. (If you have a succinct hypothesis this is one slide you *might* be able to get away with putting it up in full in text form and reading it exactly, but only if the rest of the presentation is pretty clean).

#### Be sure to lay out each analysis clearly before talking about its results

If you say what results you got before you describe the pipeline, data, etc people won't understand what's going on.

Here’s a neat trick to make complex results easier to understand: Make a diagram / cartoon of how you EXPECTED your data to look under EACH competing hypothesis. Make these in roughly the same format as the data you will present right afterwards.  You can then say something like, “So what really happened?” and then reveal the true data. Then all your viewers have to do is match the real data to one of the patterns (or see they didn’t match any of them) to get you point when you reveal the real results.  The result is either a) it’s clear which pattern it matched or b) laughter (which you can play up) if the real results were wildly different from any of your hypotheses. If b) be sure to describe how you adjusted your hypotheses and ran other tests to follow up (or at least say how you would do that/ what you think is going on).
If you show a figure, you MUST explain what the x-axis, y-axis, points, colors and series represent, and also point to the specific features of the graph that you draw on in formulating any conclusions downstream. It seems like people will figure it out, but in general they won't. I usually say something like (totally hypothetical example): 'I've plotted the results of this analysis here. The x-axis is time, and the y-axis is microbiome diversity.  The different colors represent the different treatments. Blue is control, and red is high-temperature. You can see that initially these two treatments look similar, but after 2 weeks of exposure, the high-temperature microbiome shows reduced diversity.  That difference is significant at p<0.05 by permutation test for the 2 week timepoint, and not significant for 1 week.  So we think that temperature affects the microbiome, but it takes somewhere between one and two weeks for the effects to sink in."

#### Slide transitions can make your presentation faster and clearer. 

What I mean here is that if you have 3 hypotheses or 3 series on a graph, consider bringing them in onto the slide one by one.  Then you know exactly what to say/which results to explain each time you hit next time, constraining your performance.  To prevent slide elements from jumping around, make a final slide with all the elements, then duplicate the final slide three times. Delete the elements you want to bring in from the 1st and 2nd copy of the slide.

The other kind of transitions (wipes, slides, dissolves, whatever) are mostly distracting. Use judiciously where appropriate if you like, but consider carefully if they add or subtract from clarity (usually they subtract).  Places where they are occasionally useful is as a metaphor (things like zooming in on a place on a map where one slide is the overview and one is about that place)

#### Use a consistent visual style

You want your viewers to focus on your results, not on random decorations on the slide or (worse) inconsistent decorations/fonts/arrangements between slides.  Just like in Tufte’s recommendations for avoiding ‘chartjunk’ in graphs, consider removing extraneous graphics to make your key results and diagrams pop. If you do include a decoration, make sure to use it consistently, and make it serve a purpose, like supporting the slide header or the like.

#### Answer audience questions

As you read papers, think about how you will use them to answer key questions. If you verbally cite a paper the information you will want to covey is:
First author, year, journal, lab of senior (last) author.
So a good verbal citation would be something like:
“So Jeff Gordon’s lab at Washington University has addressed the question of microbiome evolution across diverse animals. As I remember from the 2012 Muegge et al., paper in Science paper ….”
If you don’t remember all that, try to at least remember the author, year and journal
If not, try to at least provide one of the above pieces of information.
NOTE: this is tough at first, but gets easier as you use the citations in your writing, and  especially as you start to meet many of the key players at conferences and so on.
